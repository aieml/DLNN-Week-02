````python
from sklearn.datasets import load_iris
#load_iris is the function used to load the iris dataset

dataset=load_iris()
#calling the load_iris function and loading the dataset

data=dataset.data #features
target=dataset.target #labels

print(data[:5])
print(target[:5])
````

````python
from keras.layers import Dense #densly connected layers
from keras.models import Sequential

model=Sequential() #model is an empty NN
#1st hidden layer (dense type-fully connected)
model.add(Dense(8,input_dim=4,activation='relu'))
#2nd Hidden layer
model.add(Dense(8,input_dim=8,activation='relu'))
#final layer
model.add(Dense(3,input_dim=8,activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
````

````python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,
                                                              test_size=0.1)
````

````python
from keras.utils import np_utils

new_train_target=np_utils.to_categorical(train_target)
print(new_train_target[:5])
````

````python
history=model.fit(train_data,new_train_target,epochs=1000)
#model.fit-to train the NN
````

##### We trained the NN for 100 epochs, try with 1000 epochs and see the improvement of the accuracy and the decrement of the loss 
After the NN is trained, the loss and accuracy can be visualized with matplotlib. All the training information is saved in the 
```history``` object. refer the code ```histroy=model.fit(train_data,new_train_target,epochs=100)``` 

````python
from matplotlib import pyplot as plt

plt.plot(history.history['loss'])
plt.xlabel('# epochs')
plt.xlabel('loss')
plt.show()
````

````python
plt.plot(history.history['accuracy'])
plt.xlabel('# epochs')
plt.xlabel('accuracy')
plt.show()
````

#### Testing NN model with testing data

we kept ```test_data``` and ```test_target``` in order to test the NN model using ```model.predict()```. There are 15 samples. Firstly we can apply the ```test_data``` and get the ```predicted_target```. Note that ```test_target``` contains the actual targets corresponds to ```test_data```

````python
predicted_target=model.predict(test_data)
print(predicted_target)
````

Note that when the predicted_target is printed, it shows a 15x3 array. each columns represent the probabilities predicted by the NN for the given ```test_data```. Using ```np.argmax(predicted_target,axis=1)``` the corresponding index (label) can be obtained. This gives the index corresponding to the maximum value of each row, which is equal to the label having the maximum probability.```axis=1``` is given to get the maximum value of each row. [read more at numpy official documents](https://docs.scipy.org/doc/numpy/reference/generated/numpy.argmax.html)

````python
import numpy as np

print('Predicted Labels:',np.argmax(predicted_target,axis=1))
print('Actual Labels:',test_target)
````

It can be seen that almost all the predicted labels are matched with the actaul lables. This is what even a simple NN can do :D
